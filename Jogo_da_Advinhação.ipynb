```python
import random as rd

def gera_numero():
    numero_aleatorio = rd.randint(0,100)
    return numero_aleatorio
    
def mensagem_boas_vindas():
    print('Bem vindo ao jogo da Advinhação!')
    print('================================')
    
def rodadas():    
    total_rodadas = int(input('Deseja jogar quantas rodadas?'))
    return total_rodadas

def jogar_novamente():

    jogar = input('Deseja jogar novamente? S - para sim ou N - para não\n')
    jogar = jogar.casefold()
    
    if jogar == 's':
        jogar_advinhacao()
    else:
        exit()

def verifica_numero(numero_aleatorio, total_rodadas):
    rodada = 1
    numero_aleatorio = numero_aleatorio
    #Para ver o número escolhido
    #print(numero_aleatorio)
    
    while rodada <= total_rodadas:
        print(f'Rodada {rodada}\n')
        chute = int(input('Digite um número:'))
        
        if chute == numero_aleatorio:
            print('Parabéns você acertou\n')
            break
        else:
            if chute > numero_aleatorio:
                print('O número chutado é maior do que o número secreto\n')
            else:
                print('O número chutado é menor do que o número secreto\n')
        
        if rodada == total_rodadas:
            print(f'Fim de jogo! O número secreto era:{numero_aleatorio}')
            jogar_novamente()
        
        rodada+=1
        
        
def jogar_advinhacao():
    mensagem_boas_vindas()
    verifica_numero(gera_numero(),rodadas())


jogar_advinhacao()
```
